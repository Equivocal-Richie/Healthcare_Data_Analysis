<h1><b>Business Understanding</b></h1>
<p style="font-size:16px;">
    This analysis aims to gain insights into the performance and trends of healthcare providers based on outpatient and inpatient charges across three years (2011-2013). The findings will be valuable for understanding cost patterns, identifying affordable providers, and analyzing prominent diagnoses by state.
</p>
<p style="font-size:16px;">
   <b>Key Goals:</b><br>
    <ol style="font-size:16px;">
        <li> Compare Provider Charges: Determine and compare average outpatient (including copayments) and inpatient Medicare payments across providers for each year.</li>
        <li>Track Charge Trends: Conduct a time series analysis to understand how the average charge for different outpatient services (APC codes) varies over time for each provider.</li>
        <li>Identify Affordable Providers: Analyze the geographical distribution of affordable providers offering emergency services across different states.</li>
        <li>Analyze Inpatient Diagnoses: Explore the top 10 most prevalent inpatient diagnoses categorized by ICD10 chapters and blocks, grouped by state.</li>
    </ol>
</p>

```python
# Identify columns with missing values
missing_values = hospital_info.columns[hospital_info.isnull().any()]

# Fill missing values in categorical columns with 'Not Available'
for column in missing_values:
    if hospital_info[column].dtype == 'object':
        hospital_info[column].fillna('Not Available', inplace=True)

# Fill missing values in numerical columns with the mean
for column in missing_values:
    if hospital_info[column].dtype == 'float64':
        hospital_info[column].fillna(hospital_info[column].mean(), inplace=True)

# Verify if missing values have been handled
print(hospital_info.isnull().sum().sum())
```

<h1><b>4. Data Analysis</b></h1>

<h2><b>4.1: Average Outpatient and Inpatient Medicare Payments Ranked per Provider for each Year</b></h2>

```python
# 2011 Analysis
# Calculate average outpatient payments per provider
outpatient_avg_payments = outpatient_2011.groupby('provider_id')['average_total_payments'].mean()

# Calculate average inpatient Medicare payments per provider
inpatient_avg_medicare_payments = inpatient_2011.groupby('provider_id')['average_medicare_payments'].mean()

# Combine results and rank
combined_payments = pd.concat([outpatient_avg_payments, inpatient_avg_medicare_payments], axis=1)
combined_payments.columns = ['avg_outpatient_payment', 'avg_inpatient_medicare_payment']
combined_payments['total_avg_payment'] = combined_payments.sum(axis=1)
combined_payments.sort_values(by='total_avg_payment', ascending=False, inplace=True)

# Prepare data for visualization 
top_10_providers = combined_payments.head(10)

# Create a Seaborn bar chart
fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(x='provider_id', y='total_avg_payment', data=top_10_providers, ax=ax, color="black")

# Customize the plot
ax.set_title('Top 10 Providers by Total Average Payment (2011)')
ax.set_xlabel('Provider ID')
ax.set_ylabel('Total Average Payment')
ax.tick_params(bottom=False, rotation=45)  # Rotate x-axis labels
# ax.legend(title='Payment Type')  # Add a legend for clarity

# Enhance interactivity with hover data 
for bar in ax.containers[0]:
    ax.annotate(f"{bar.get_height() :.2f}", xy=(bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                ha='center', va='bottom', fontsize=10)  # Show individual values on hover

plt.tight_layout()
plt.show()
```

```python
# Combine inpatient data for all years (2011-2013)
inpatient_data = pd.concat([inpatient_2011, inpatient_2012, inpatient_2013])

# Group by state and ICD category, then count occurrences
state_icd_counts = inpatient_data.groupby(['provider_state', 'icd_category'])['icd_category'].count().unstack(fill_value=0)

# Get the top 10 ICD categories for each state (by sum)
top_10_per_state = state_icd_counts.apply(pd.Series.nlargest, n=10)

# Reshape the DataFrame for plotting (optional)
top_10_per_state = top_10_per_state.stack().reset_index(name='count')

# Create a Seaborn bar plot
sns.set_style("whitegrid")  # Set background style

fig, ax = plt.subplots(figsize=(12, 6))  # Adjust figure size

# Plot the top 10 diagnoses for each state
sns.barplot(x='provider_state', y='count', hue='icd_category', data=top_10_per_state)

# Rotate x-axis labels for readability (optional)
plt.xticks(rotation=45, ha='right')

# Add labels and title
plt.xlabel('State')
plt.ylabel('Count')
plt.title('Top Inpatient Diagnoses by State (2011-2013)')

# Adjust legend size (optional)
plt.legend(title='ICD Category', loc='upper left', bbox_to_anchor=(1, 1), fontsize=8)  # Smaller legend, upper left position

plt.tight_layout()
plt.show()
```
